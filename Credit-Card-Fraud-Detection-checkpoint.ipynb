```python
def remove_outliers_iqr(df):
    # Copy the DataFrame to avoid modifying the original data
    df_cleaned = df.copy()

    # List to store columns that are numerical
    numerical_columns = df_cleaned.select_dtypes(include=['number']).columns

    # Iterate over each numerical column
    for column in numerical_columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df_cleaned[column].quantile(0.25)
        Q3 = df_cleaned[column].quantile(0.75)
        # Calculate IQR (Interquartile Range)
        IQR = Q3 - Q1
        # Define bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Filter out the outliers
        df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]

    return df_cleaned
```

```python
X = df.drop('is_fraud',axis=1)
y=df.is_fraud
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,y,test_size=.20,stratify=y,random_state=7)
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
lr_clf = LogisticRegression(max_iter=2000,random_state=7,verbose=True)
lr_clf.fit(x_train,y_train)
```

```python
lr_pred = lr_clf.predict(x_test)
print(f"Lr_clf accuracy is {accuracy_score(y_test,lr_pred)}")
```

```python
dt_clf = DecisionTreeClassifier(random_state=7)
dt_clf.fit(x_train,y_train)
```

```python
dt_pred = dt_clf.predict(x_test) 
print(f"dt_clf accuracy is {accuracy_score(y_test,dt_pred)}")
```

```python
n = [3,5,7,9,11,13,15]
scores = []
for i in n:
    knn_clf = KNeighborsClassifier(n_neighbors=i)  
    knn_clf.fit(x_train,y_train)
    knn_pred = knn_clf.predict(x_test)
    scores.append(accuracy_score(y_test,knn_pred))
print(scores)
```

```python
rf_clf = RandomForestClassifier(random_state=7)
rf_clf.fit(x_train,y_train)
```

```python
rf_pred = rf_clf.predict(x_test)
print(f"rf_clf accuracy is {accuracy_score(y_test,rf_pred)}")
```

```python
svm_clf = SVC(kernel = 'linear',random_state=7)
svm_clf.fit(x_train,y_train)
```

```python
svm_pred = svm_clf.predict(x_test)
print(f"svm_clf accuracy is {accuracy_score(y_test,svm_pred)}")
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(17,)),  # Input layer with 9 features
    tf.keras.layers.Dense(64, activation='relu'),  # Hidden layer
    tf.keras.layers.Dense(32, activation='relu'),  # Another hidden layer
    tf.keras.layers.Dense(2, activation='softmax')  # Output layer for binary classification
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # For multi-class classification
    metrics=['accuracy']
)

# Train the model
model.fit(x_train, y_train, epochs=100, validation_split=0.2)

# Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_acc}")

# Make predictions
predictions = model.predict(x_test)
print(f"Predicted label for the first test case: {tf.argmax(predictions[0])}")
```
